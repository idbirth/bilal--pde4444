## This lab has two tasks in which we will use the K-means with Elbow method on a given dataset with two features; Weight and Height. 
### The dataset is provided in the “Clustering.csv” file


```python
# Import the libraries:

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import metrics
from sklearn.cluster import KMeans # for kmeans clustering

```

```python
#read the datset and plot it to see the data,
# you can use the plt.scatter function to plot the data directly from the datafreame
data = pd.read_csv('Clustering.csv')

plt.figure(figsize=(5,5))
plt.scatter(data["Weight"],data["Height"])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Data Distribution')
plt.show()
```

```python
# apply the elbow method
# the objective of the Elbow method is to know the best number of clusters
# that will maximize the variance between different clusters and minimise the variance 
# between daya samples in the same cluster (inertia)
data = pd.read_csv('Clustering.csv')
X = data
variance = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    variance.append(kmeans.inertia_)

fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 20), variance)
plt.grid(True)
plt.title('Elbow curve')
plt.show()
```

```python
#create the clustering model using Kmeans
data = pd.read_csv('Clustering.csv')
Num_Cluters = 3
kmeansModel = KMeans(n_clusters=Num_Cluters) #based on the elbow method


#training k-means model
kmeansModel.fit(data)

#predictions from kmeans
pred = kmeansModel.predict(data)
print(pred) #pred is an array contains the cluster label for each data sample

#plotting results
frame = pd.DataFrame(data)
frame['cluster'] = pred
frame.columns = ['Weight', 'Height', 'cluster']
color=['blue','green','cyan', 'black']
for k in range(0,Num_Cluters):
    data = frame[frame["cluster"]==k]
    plt.scatter(data["Weight"],data["Height"],c=color[k])
plt.show()
```
